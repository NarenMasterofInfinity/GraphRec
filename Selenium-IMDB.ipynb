```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import csv

options = Options()
options.headless = True

driver = webdriver.Remote(
    command_executor="http://localhost:4444/wd/hub",
    options=options
)

all_movies = []

base_url = "https://www.imdb.com/search/title/?groups=top_1000&start={}&ref_=adv_nxt"

for start in range(1, 1001, 50):  
    print(f"Scraping page starting at movie {start}")
    driver.get(base_url.format(start))
    time.sleep(2)

    movie_blocks = driver.find_elements(By.CSS_SELECTOR, ".lister-item.mode-advanced")

    for movie in movie_blocks:
        try:
            title = movie.find_element(By.CSS_SELECTOR, "h3.lister-item-header a").text
        except NoSuchElementException:
            title = "N/A"
        
        try:
            p_tags = movie.find_elements(By.CSS_SELECTOR, ".lister-item-content > p")
            summary = "N/A"
            if len(p_tags) >= 2:
                summary = p_tags[1].text.strip()
        except NoSuchElementException:
            summary = "N/A"

        all_movies.append([title, summary])

    time.sleep(1)

driver.quit()

# Save to CSV
with open("imdb_top_1000.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Plot Summary"])
    writer.writerows(all_movies)

print(f"Scraped {len(all_movies)} movies and saved to imdb_top_1000.csv")
```
